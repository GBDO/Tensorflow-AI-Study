- 연산 과정에 부동소수가 있다면 결과는 부동소수
- /(나눗셈)은 무조건 부동소숫값 반환

### Floating-Point : 실수형(부동소수형) 자료를 저장. (크기제한 없음)
- ex) 1.0, -5.3, 3.1415, 3e5 ...
- 위의 예시처럼 e(지수표기법) 사용가능. e는 *10<sup>(뒤의 수)</sup>로 해석
- 정수와의 연산에서는 정수가 실수로 자동으로 형변환됨
- 연산 결과는 항상 실수

````python
print(type(1.0), type(-5.3), type(3.1415), type(3e5))
````

````python
5 + 3.14
````

### Bool : 논리(참거짓) 자료를 저장. True, False만 저장가능

````python
print(type(True), type(False))
````

````python
True and True, True and False, False and True, False and False
````

````python
True or True, True or False, False or True, False or False
````

````python
not True, not False
````

````python
5 == 5, 5 <= 2, 5 >= 2, 5 < 2, 5 > 2, 5 != 2
````

#### if, elif, else 조건문
```python
if 조건1:
    실행내용
elif 조건2:
    실행내용
else:
    실행내용
```

````python
score = 88

if score > 80:
    print("pass")
    
else:
    print('fail')
````

````python
if score > 90:
    print("A")
elif score > 80:
    print("B")
elif score > 70:
    print("C")
elif score > 60:
    print("D")
else:
    print("F")
````
